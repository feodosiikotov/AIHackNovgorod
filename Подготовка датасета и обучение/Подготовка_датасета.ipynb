```python
from docx import Document
from docx.opc.exceptions import PackageNotFoundError
```

```python
import os
import re
import pandas as pd
```

```python
base = 'DataSet_Razmetra'
ne = 0
d = 0
wrong_samples = []
true_samples = []
cls = 0
n = 0
for f1 in os.listdir(base):
  for f2 in os.listdir(f'{base}/{f1}'):
    for f3 in os.listdir(f'{base}/{f1}/{f2}'):
      try:
        for d0 in os.listdir(f"{base}/{f1}/{f2}/{f3}"):
          if d0.startswith('Edit'):
            d = d0
            break
        doc1 = Document(f'{base}/{f1}/{f2}/{f3}/{d}/Edition_Text.docx')
        n += 1
        has_color = False
        for para in doc1.paragraphs:            
          for run in para.runs:
            if run.font.highlight_color in (6, 7):
              has_color = True
              cls += 1
              break
          if has_color:
            wrong_samples.append([para.text, f2])
          else:
            true_samples.append([para.text, f2])
          
      except (PackageNotFoundError, NotADirectoryError):
        pass
```

```python
pd.DataFrame(wrong_samples).to_csv('wrong_samples_color.csv')
pd.DataFrame(true_samples).to_csv('true_samples_color.csv')
```
